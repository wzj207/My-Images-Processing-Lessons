## Image are numpy arrays
- __build a grayscale image as a 2D array.__

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

random_img = np.random.random([500,500,])

plt.imshow(random_img,cmap='gray')
plt.colorbar()
```

- __"real-world" images__

```python
from skimage import data

coins = data.coins()

print('Type:',type(coins))
print('dtype:',coins.dtype)
print('shape:',coins.shape)

plt.imshow(coins,cmap='gray')
```

- __color image__

```python
cat = data.chelsea()

print('Shape:',cat.shape)
print('Values min/max:',coins.min(),coins.max())

plt.imshow(cat,cmap='gray')
# plt.axis('off')
```

```python
cat[10:110,10:110,:] = [255,0,0]
plt.imshow(cat)
```

## __Visualizing image channel__

```python
# --- read the image ---

image = plt.imread('./images/Bells-Beach.jpg')

# --- assign each color channel to a different variable ---

r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

# --- display the image and r,g,b channels ---

f,axes = plt.subplots(1,4,figsize=(16,5))

for ax in axes:
    ax.axis('off')
    
(ax_r,ax_g,ax_b,ax_color) = axes

ax_r.imshow(r,cmap='gray')
ax_r.set_title('red channel')

ax_g.imshow(g,cmap='gray')
ax_g.set_title('green channel')

ax_b.imshow(b,cmap='gray')
ax_b.set_title('blue channel')

# ax_color.imshow(image,cmap='gray')
# ax_color.imshow(np.dstack((r,g,b)))
ax_color.imshow(np.stack((r,g,b),axis=2))
ax_color.set_title('all channel')
```
