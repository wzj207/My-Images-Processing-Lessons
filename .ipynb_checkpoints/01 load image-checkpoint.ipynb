```python
from skimage import io
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
```

## load a single image

```python
balloon = io.imread('./images/balloon.jpg')

f,ax = plt.subplots()

ax.imshow(balloon,cmap='gray')
ax.axis('off')
ax.set_title(os.path.basename('./images/balloon.jpg'))
```

```python
height,width,channels = balloon.shape
print('height:',height,'\nwidth:',width,'\nchannels:',channels)
print()
print('min:',balloon.min(),'\nmax:',balloon.max(),'\nmean:',balloon.mean())
print()
print('size:',balloon.size)
```
