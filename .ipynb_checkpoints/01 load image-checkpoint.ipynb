```python
from skimage import io
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
```

## load a single image

```python
balloon = io.imread('./images/balloon.jpg')

f,ax = plt.subplots()

ax.imshow(balloon,cmap='gray')
ax.axis('off')
ax.set_title(os.path.basename('./images/balloon.jpg'))
```

```python
height,width,channels = balloon.shape
print('==================== basic information ============================')
print('Type:',type(balloon))
print('dtype:',balloon.dtype)
print()
print('height:',height,'\nwidth:',width,'\nchannels:',channels)
print()
print('min:',balloon.min(),'\nmax:',balloon.max(),'\nmean:',balloon.mean())
print()
print('size:',balloon.size)
print('==================== basic information ============================')
```

## load multiple images

```python
ic = io.ImageCollection('./images/*.png;./images/*.jpg')

print('Type:',type(ic))

ic.files[:5]
```

```python
import os

f,axes = plt.subplots(nrows=3,ncols=len(ic) // 3 + 1,figsize=(20,5))

axes = axes.ravel()

for ax in axes:
    ax.axis('off')

for i,image in enumerate(ic):
    axes[i].imshow(image,cmap='gray')
    axes[i].set_title(os.path.basename(ic.files[i]))
plt.tight_layout()
```
